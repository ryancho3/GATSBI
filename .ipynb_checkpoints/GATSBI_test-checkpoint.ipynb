```python
DATA = api.get_bars("SPY", TimeFrame.Day, start = "2021-06-01", adjustment='raw').df
DATA['EMA1'] = DATA['close'].ewm(span=10, adjust=False).mean()
DATA['EMA2'] = DATA['close'].ewm(span=15, adjust=False).mean()
plt.figure(figsize=(10,5))

print(DATA)
```

```python
DATA['Cross'] = 0.0
DATA['Cross'] = np.where(DATA['EMA1'] > DATA['EMA2'], 1.0, 0.0)
DATA['Signal'] = DATA['Cross'].diff()

map_dict = {-1.0: 'sell', 1.0: 'buy', 0.0: 'none'}
DATA["Signal"] = DATA["Signal"].map(map_dict)

print(DATA)
```

```python
# Plot stock price data
plt.plot(DATA['close'])
plt.plot(DATA['EMA1'])
plt.plot(DATA['EMA2'])

# Plot ‘buy’ signals
plt.plot(DATA[DATA['Signal'] == 'buy'].index,
         DATA['EMA2'][DATA['Signal'] == 'buy'],
         '^', markersize=8, color='g', label='buy')

# Plot ‘sell’ signals
plt.plot(DATA[DATA['Signal'] == 'sell'].index,
         DATA['EMA2'][DATA['Signal'] == 'sell'],
         'v', markersize=8, color='r', label='sell')

plt.xlabel("Date")
plt.ylabel("Close Price ($)")
plt.legend()
plt.show()
```

```python
transaction_type = ['buy', 'sell']

print(DATA[DATA['Signal'].isin(transaction_type)])
```

```python
class Agent:
    def __init__(self, data, g):
        self.g = g
        self.data = data
    def randomize(self):
        self.g = []
        for i in range(2):
            g.append(random.randint(2,30))
            
    def execute(self):
        self.data['EMA1'] = self.data['close'].ewm(span=self.g[0], adjust=False).mean()
        self.data['EMA2'] = self.data['close'].ewm(span=self.g[1], adjust=False).mean()
        self.data['cross'] = 0.0
        self.data['cross'] = np.where(self.data['EMA1'] > self.data['EMA2'], 1.0, 0.0)
        self.data['signal'] = self.data['cross'].diff()

        map_dict = {-1.0: 'sell', 1.0: 'buy', 0.0: 'none'}
        self.data["signal"] = self.data["signal"].map(map_dict)
        
        return self.data
    
    def calculate_fitness(self):
        self.transaction_data = self.data[self.data['signal'].isin(transaction_type)]
        signal = self.transaction_data['signal'].tolist()
        close = self.transaction_data['close'].tolist()
        self.fitness = 0.0
        for i in range(1,len(signal)):
            if signal[i-1] == 'buy' and signal[i] == 'sell':
                self.fitness = self.fitness + (close[i] - close[i-1])
                print('buy-sell performance: ' + str((close[i] - close[i-1])))
            elif signal[i-1] == 'sell' and signal[i] == 'buy':
                self.fitness = self.fitness + (close[i-1] - close[i])
                print('sell-buy performance: ' + str((close[i-1] - close[i])))
        return self.fitness
    
    def plot(self):
        plt.plot(self.data['close'])
        plt.plot(self.data['EMA1'])
        plt.plot(self.data['EMA2'])

        plt.plot(self.data[self.data['signal'] == 'buy'].index,
                 self.data['EMA2'][self.data['signal'] == 'buy'],
                 '^', markersize=8, color='g', label='buy')

        # Plot ‘sell’ signals
        plt.plot(self.data[self.data['signal'] == 'sell'].index,
                 self.data['EMA2'][self.data['signal'] == 'sell'],
                 'v', markersize=8, color='r', label='sell')

        plt.xlabel("Date")
        plt.ylabel("Close Price ($)")
        plt.legend()
        plt.show()

    
        
                
                
```

```python
input = api.get_bars("SPY", TimeFrame.Day, start = "2021-06-01", adjustment='raw').df
input = input.drop(['open', 'high', 'low', 'volume', 'trade_count', 'vwap'], axis=1)
```

```python
test_agent = Agent(input, [5,10])
print(test_agent.execute())
print(test_agent.calculate_fitness())
```
